```python
import numpy as np
import matplotlib.pyplot as plt
from cued_sf2_lab.familiarisation import load_mat_img, plot_image
import IPython.display
```

```python
X0, _ = load_mat_img('lighthouse.mat', img_info='X')
X1, _ = load_mat_img('bridge.mat', img_info='X')
X2, _ = load_mat_img('flamingo.mat', img_info='X')
X3, _ = load_mat_img('house.mat', img_info='M')
X4, _ = load_mat_img('face.mat', img_info='M')
```
